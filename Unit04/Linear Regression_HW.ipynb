## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
regr = LinearRegression()
#注意轉換x得格式1D->2D
x_train = x_train.reshape([-1,1])
#將x,y資料導入LinearRegression演算法做訓練
regr.fit(x_train,y_train)
#獲得預測值
y_pred = regr.predict(x_train)
#作圖
plt.scatter(x_train,y_train)
plt.scatter(x_train, y_pred, c = 'r')
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
x_test = x_test.reshape([-1,1])
y_test_pred = regr.predict(x_test)
# plt.scatter(x_test, y_test_pred)
```

```python
w = round(float(regr.coef_),1)
b = round(regr.intercept_,1)

print('斜率：', w)
print('截距：', b)
```

2. 並計算出測試階段的MSE

```python
mse = np.sum(y_test_pred-y_test) ** 2 / len(y_test)
round(mse,2)
```

3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_test.PNG)

```python
plt.scatter(x_test, y_test,label = 'Ground Truth')
plt.scatter(x_test, y_test_pred, c = 'r', label = 'Predict')
plt.legend()
plt.title(label = 'f(x)=1.6x+0.2,mse=0.52')
plt.show()
```
